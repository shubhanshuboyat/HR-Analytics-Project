# Employee Attrition Distribution

```python
plt.figure(figsize=(12,6))

sns.set_style=("dark")

ax = sns.countplot(
    data=df,
    x="attrition",
    hue="attrition",
    palette="Set2",
    legend=False
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Employee Attrition Distribution",
          fontsize=14,
          weight="bold"
)

plt.xlabel("Attrition Status")
plt.ylabel("Number of Employee")
plt.grid(axis="y",linestyle="--")

plt.tight_layout()

plt.savefig("images/Attrition_distribution.png",
            bbox_inches="tight",
            dpi=300
)

plt.show()
```

# Business Insight
The company has good employee retention but reducing can further improve workforce stability and perfomance.
# Key Finding
Most employee stay with the company , while only a small number leave the organization.

# Department Wise Attrition % Analysis

```python
attrition_rate = (
    df.groupby("department")["attrition"]
    .apply(lambda x: (x == "Yes").mean()*100)
    .sort_values(ascending=False)
)

attrition_rate
```

```python
plt.figure(figsize=(12,5))

ax = sns.barplot(
    x=attrition_rate.index,
    y=attrition_rate.values,
    palette="Set2"
)

for contriner in ax.containers:
    ax.bar_label(container , fmt="%.2f")

plt.title("Department Wise Attrition Rate %",
          fontsize=14,
          weight="bold")

plt.xlabel("Department")
plt.ylabel("Number of Employee")

plt.tight_layout()

plt.savefig("images/Department_wise.png",
            bbox_inches="tight",
            dpi=300)

plt.show()
```

# Business Insight
- Sales department has the hightest employee attrition compared to other departments.
- This suggests that workload,targets or employee satisfaction mayy need improvemnet in this area.

# key Finding
- Employee is sales leave the company more frequently than other department.
- High turnover may affect productivity and business performance.

# Job Role Chart

```python
top_roles = (
    df[df["attrition"] == "Yes"]["jobrole"]
    .value_counts()
    .head()
)
top_roles
```

```python
plt.figure(figsize=(12,6))

ax = sns.barplot(
    x=top_roles.values,
    y=top_roles.index,
    palette="Set2"
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Top job Roles with Highest Attrition",
          fontsize=14,
          weight="bold")

plt.xlabel("Nmuber of Employee")
plt.ylabel("Fob Role")

plt.tight_layout()

plt.savefig("images/job_role_attrition.png",
            bbox_inches="tight",
            dpi=300)

plt.show()
```

# Business Insight
- Some job roles experience significantly higher employee attrition than others.
- HR should focus retention efforts on these critical roles.
# Key Finding
- The highest attrition is concentrated in a few specific job roles.
- Tangeted retention programs can help reduce employee turnover.

```python
plt.figure(figsize=(12,5))

ax = sns.countplot(
    data=df,
    x="overtime",
    hue="attrition",
    palette="Set2"
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Overtime vs Attrition",
          fontsize=14,
          weight="bold")

plt.xlabel("Overtime")
plt.ylabel("Number of Employee")

plt.tight_layout()

plt.savefig("images/Overtime_vs_Attrition.png",
            bbox_inches="tight",
            dpi=300)

plt.show()
```

- High worked and work life imbalance may be major reasons behind employee turnover.
# Key Finding
- Overtime employees show an attrition rate of around 30%, compared to only 10% for non-overtime employees.
-Overtime is one of the strongest factors linked to employee attrition.

# Monthly Income vs Attrition💰

```python
df["income_group"] = pd.cut(
    df["monthlyincome"],
    bins=[0,5000,10000,15000,20000],
    labels=["Lows","Medium","High","Very Hight"]
)
df["income_group"].value_counts()
```

```python
plt.figure(figsize=(12,5))

ax = sns.countplot(
    data=df,
    x="income_group",
    hue="attrition",
    palette="Set2"
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Monthly Income vs Attrition",
          fontsize=14,
          weight="bold")

plt.xlabel("Income Group")
plt.ylabel("Number of Employees")

plt.tight_layout()

plt.savefig("images/Income_vs_Attrition.png",
            bbox_inches="tight",
            dpi=300)

plt.show()
```

```python
pd.crosstab(
    df["income_group"],
    df["attrition"]
)
```

# Business Indight
- Employee in the Low Income group show the highest attrition compared to other salary groups.
-Increasing compensation and employee benefits may help reduce employee turnover.
# Key Finding
- Attrition decreases as employee income increases.
- Very High income employees have the lowest attrition rate in the company.

# Age Group vs Attrition

```python
df["age_group"] = pd.cut(
    df["age"],
    bins=[18,25,35,45,60],
    labels=["18-25","26-35","36-45","46-60"]
)
df["age_group"].value_counts()
```

# Business Insight
- Mid career employees are more likely to leave the organization.
- career growth opportunities can help improve retenttion.
# Key Finding
- Employees aged 26-35 have the highest attrition.
- Attrition is lower among older age groups.
